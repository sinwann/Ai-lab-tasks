Depth First Search (DFS) Task 01

```python
# Depth First Search Implementation

graph = {
    0: [1, 2],
    1: [0, 3],
    2: [0, 4],
    3: [1],
    4: [2]
}

visited = set()

def dfs(node):
    if node not in visited:
        print(f"Visited Node: {node}")
        print(f"Adjacent Nodes of {node}: {graph[node]}")
        visited.add(node)

        for neighbor in graph[node]:
            dfs(neighbor)

print("DFS Traversal Starting from Node 0:\n")
dfs(0)
```

Breadth First Search (BFS) Task 02

```python
# Breadth First Search Implementation

from collections import deque

graph = {
    0: [1, 2],
    1: [0, 3],
    2: [0, 4],
    3: [1],
    4: [2]
}

visited = set()
queue = deque()

def bfs(start_node):
    visited.add(start_node)
    queue.append(start_node)

    while queue:
        node = queue.popleft()
        print(f"Visited Node: {node}")
        print(f"Adjacent Nodes of {node}: {graph[node]}")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

print("BFS Traversal Starting from Node 0:\n")
bfs(0)
```

```python
# Breadth First Search Implementation

from collections import deque

graph = {
    0: [2, 3],
    1: [2],
    2: [0, 1, 4],
    3: [0],
    4: [2]
}

visited = set()
queue = deque()

def bfs(start_node):
    visited.add(start_node)
    queue.append(start_node)

    while queue:
        node = queue.popleft()
        print(f"Visited Node: {node}")
        print(f"Adjacent Nodes of {node}: {graph[node]}")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

print("BFS Traversal Starting from Node 0:\n")
bfs(0)
```
