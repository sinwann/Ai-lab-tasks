append()

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")

print(fruits)
```

clear()

```python
numbers = [1, 2, 3, 4]
numbers.clear()

print(numbers)
```

copy()

```python
list1 = ["AI", "ML", "DL"]
list2 = list1.copy()

print("Original List:", list1)
print("Copied List:", list2)
```

count()

```python
marks = [85, 90, 75, 90, 88, 90]
count_90 = marks.count(90)

print("Number of students with 90 marks:", count_90)
```

extend()

```python
list1 = ["CS", "AI"]
list2 = ["ML", "DS"]

list1.extend(list2)
print(list1)
```

index()

```python
subjects = ["Math", "Physics", "AI", "Chemistry"]
index_ai = subjects.index("AI")

print("Index of AI:", index_ai)
```

insert()

```python
students = ["Ali", "Ahmed", "Sara"]
students.insert(1, "Ayesha")

print(students)
```

pop()

```python
cities = ["Lahore", "Karachi", "Islamabad"]
removed_city = cities.pop(1)

print("Removed City:", removed_city)
print("Remaining Cities:", cities)
```

remove()

```python
colors = ["red", "blue", "green", "blue"]
colors.remove("blue")

print(colors)
```

reverse()

```python
numbers = [1, 2, 3, 4, 5]
numbers.reverse()

print(numbers)
```

sort()

```python
values = [45, 12, 89, 33]
values.sort()

print(values)
```

Descending Order:

```python
values.sort(reverse=True)
print(values)
```
